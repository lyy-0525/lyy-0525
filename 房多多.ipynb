```python
import requests
import time
import json
from lxml import etree
```

```python
heads={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
```

```python
def get_info(url):
    r=requests.get(url,headers=heads)
    home=etree.HTML(r.text)
    mes=home.xpath('//*[@id="root"]/main/div[3]/div[1]/ul/li')
    #file=open('homework_result.txt','a', encoding='utf-8')
    for me in mes:
        #item={}
        try:
            
            name="-".join(me.xpath('.//div/h4/a/span/text()')[0].split())
            address="-".join([me.xpath('.//div/p[2]/a[1]/text()')[0],me.xpath('.//div/p[2]/a[2]/text()')[0],me.xpath('.//div/p[2]/a[3]/text()')[0]])
            imformation="-".join([me.xpath('.//div/p[1]/text()')[0],me.xpath('.//div/p[1]/text()')[1],me.xpath('.//div/p[1]/text()')[2]])
            sale=''.join([me.xpath('.//div/div[3]/p[1]/strong/text()')[0],"万"])#通过join函数连接
            price=me.xpath('.//div/div[3]/p[2]/text()')[0]
            print(name+","+address+","+imformation+","+sale+","+price+"\n")
            #print(item)
            with open('result.txt','a',encoding='utf-8') as f:
                f.write(name+","+address+","+imformation+","+sale+","+price+"\n")
            
        except:
            print('出现问题')
```

```python
file=open('result.txt','w+', encoding='utf-8')
file.write("名字,地址,信息,总价,单价"+"\n")
for i in range(1,52):
    url='https://zhuhai.fangdd.com/esf/?pageNo={}'.format(i)
    get_info(url)
    print(i)
    time.sleep(1)
print("爬信息成功")
file.close()
```
